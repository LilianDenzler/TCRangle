```python
# If needed (uncomment):
# !pip install mdanalysis tqdm pandas

from pathlib import Path
import json
from typing import Any, Dict, List, Optional

import numpy as np
import pandas as pd
from tqdm import tqdm

import MDAnalysis as mda
from MDAnalysis.coordinates.PDB import PDBWriter

from trangle.new_calc import run as trangle_run

# EDIT THESE:
pdb_path = "/mnt/larry/lilian/DATA/Cory_data/A6/A6prmtop_first_frame.pdb"     # e.g. "/workspaces/.../A6prmtop_first_frame_output_chain_split_oriented.pdb"
xtc_path = "/mnt/larry/lilian/DATA/Cory_data/A6/Prod_Concat_A6_CMD.xtc"   # your trajectory file
out_root = Path("./output_trangle")    # where results will be written

# Optional slicing:
stride = 1        # use every Nth frame (1 = all frames)
start  = 0        # first frame index to process
stop   = None     # None = until the end
vis    = True     # pass vis=True to trangle
```

```python
def write_frame_pdb(u: mda.Universe, path: Path):
    """Write the *current* frame of Universe `u` to a PDB file."""
    path.parent.mkdir(parents=True, exist_ok=True)
    with PDBWriter(str(path), multiframe=False) as w:
        w.write(u.atoms)

def run_trangle_for_frame(frame_pdb: Path, out_dir: Path, vis: bool = False):
    """Run trangle on a single PDB frame and return its result."""
    out_dir.mkdir(parents=True, exist_ok=True)
    return trangle_run(str(frame_pdb), out_dir, vis=vis)

def result_to_flat_dict(frame_idx: int, result: Any) -> Optional[Dict[str, Any]]:
    """Best-effort: turn TRangle's result into a flat dict (for CSV aggregation)."""
    if isinstance(result, dict):
        flat = {"frame": frame_idx}
        for k, v in result.items():
            if isinstance(v, (np.ndarray, list, tuple)):
                v = np.asarray(v).tolist()
            elif not isinstance(v, (str, int, float, bool)) and v is not None:
                try:
                    v = json.dumps(v, default=str)
                except Exception:
                    v = str(v)
            flat[k] = v
        return flat
    return None
```

```python
u = mda.Universe(pdb_path, xtc_path)
total = len(u.trajectory)
if stop is None:
    stop = total

frames_dir = out_root / "frames"
agg_rows: List[Dict[str, Any]] = []

print(f"Total frames: {total} | Processing frames [{start}:{stop}:{stride}]")
for i, ts in enumerate(tqdm(u.trajectory[start:stop:stride], desc="Frames")):
    frame_idx = start + i * stride

    # write this frame as a temporary PDB (under the per-frame folder so you can inspect later if needed)
    frame_out = frames_dir / f"frame_{frame_idx:06d}"
    frame_pdb = frame_out / f"frame_{frame_idx:06d}.pdb"
    write_frame_pdb(u, frame_pdb)

    # run TRangle on that single-frame PDB
    result = run_trangle_for_frame(frame_pdb, frame_out, vis=vis)
    print(result)

    # try to aggregate
    flat = result_to_flat_dict(frame_idx, result)
    if flat is not None:
        agg_rows.append(flat)
    else:
        # still persist raw result if possible
        try:
            with open(frame_out / "result.json", "w") as f:
                json.dump(result, f, default=str, indent=2)
        except Exception:
            pass

print("Done.")
```
